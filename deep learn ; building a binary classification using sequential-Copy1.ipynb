```python
import tensorflow as tf
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
x = cancer.data
y = cancer.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,stratify=y,test_size=0.3)
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10, activation = "relu", input_shape=[x_train.shape[1],],name = "hidden_1"))
model.add(Dense(20, activation = "relu", name = "hidden_2"))
model.add(Dense(10, activation = "relu", name = "hidden_3"))
model.add(Dense(10, activation = "relu", name = "hidden_4"))
model.add(Dense(1, activation = 'sigmoid', name = "output_layer"))
model.summary() # 310 weights and bais
```

```python
model.weights
```

```python
model.compile(optimizer = 'SGD', loss = 'binary_crossentropy', metrics =['accuracy'] )
```

```python
#these is training the data 
# epochs is number of iteration or number of epochs is equal to number of completing  forward path 
history = model.fit(x_train,y_train,epochs = 100,validation_data = (x_test,y_test))
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
